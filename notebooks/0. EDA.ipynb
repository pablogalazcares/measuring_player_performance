### Matches

```python
rows = []
for _ in competitionData:
    season_id = _["season_id"]
    with open("data/events/data/matches/2/"+str(season_id)+".json","r") as file:
        fileData  = file.read()
        seasonMatches = json.loads(fileData)
        
    for _ in seasonMatches:
        row = []
        for key in _.keys():
            if key == "competition":
                row.append(_[key]["competition_id"])
            elif key == "season":
                row.append(_[key]["season_id"])
            elif key == "home_team":
                row.append(_[key]["home_team_id"])
                # row.append(_[key]["home_team_name"])
            elif key == "away_team":
                row.append(_[key]["away_team_id"])
                # row.append(_[key]["home_team_name"])
            elif key == "competition_stage":
                row.append(_[key]["name"])
            elif key == "stadium":
                row.append(_[key]["name"])
            elif key == "referee":
                row.append(_[key]["name"])
            else:
                row.append(_[key])
        rows.append(row)
df_matches = pd.DataFrame(rows, columns = _.keys())
df_matches.to_csv("dataframes/df_matches.csv")
```

```python
rows = []
for match_id, season_id in zip(df_matches.match_id, df_matches.season):
    with open("data/events/data/events/"+str(match_id)+".json","r") as file:
        fileData  = file.read()
        eventsMatch = json.loads(fileData)
        
        
        for event in eventsMatch:
            row = []
            for key in columns: # event.keys():
                if key == "id":
                    idx = event[key]
                elif key == "index":
                    index = event[key]
                elif key == "period":
                    period = event[key]
                elif key == "timestamp":
                    timestamp = event[key]
                elif key == "minute":
                    minute = event[key]
                elif key == "second":
                    second = event[key]
                elif key == "type":
                    typex = event[key]["name"]
                elif key == "team":
                    team = event[key]["name"]
                elif key == "player":
                    try:
                        player = event[key]["name"]
                    except:
                        player = np.nan
                elif key == "position":
                    try:
                        position = event[key]["name"]
                    except:
                        position = np.nan
                elif key == "location":
                    try:
                        location = event[key]   
                    except:
                        location = np.nan
                elif key == "duration":
                    try:
                        duration = event[key]
                    except:
                        duration = np.nan
                elif key == "under_pressure":
                    try:
                        under_pressure = event[key]
                    except:
                        under_pressure = np.nan
                elif key == "obv_for_after":
                    obv_for_after = event[key]
                elif key == "obv_for_before":
                    obv_for_before = event[key]
                elif key == "obv_for_net":
                    obv_for_net = event[key]
                elif key == "obv_against_after":
                    obv_against_after = event[key]
                elif key == "obv_against_before":
                    obv_against_before = event[key]
                elif key == "obv_against_net":
                    obv_against_net = event[key]
                elif key == "obv_total_net":
                    obv_total_net = event[key]
                else:
                    pass    
                    
            
            row = [idx, index, match_id, period, timestamp, minute, second, typex, team, player, position, location, duration, under_pressure, obv_for_after, obv_for_before, obv_for_net, obv_against_after, obv_against_before,
                   obv_against_net, obv_total_net]
                    
            rows.append(row)

columns_events = ["id","index","match_id","period","timestamp","minute","second","type","team","player","position","location","duration","under_pressure",
                  "obv_for_after","obv_for_before","obv_for_net","obv_against_after","obv_against_before","obv_against_net","obv_total_net"]
            
df_events = pd.DataFrame(rows, columns=columns_events)
df_events.to_csv("dataframes/df_events.csv")
```
